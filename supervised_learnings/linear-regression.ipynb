```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
cData = pd.read_csv('auto-mpg.csv')
```

```python
cData.head(10)
```

```python
shape = cData.shape
print(shape)
print(f'no of data or rows: {shape[0]}')
print(f'no of columns: {shape[1]}')
```

```python
# from cData ignore car_name
cData = cData.drop('car name', axis=1)

# replace origin number with name
# 1.America, 2.Europ, 3.Asia
cData['origin'] = cData['origin'].replace({1:"america", 2:"europ", 3:"asia"})
```

```python
cData.head()
```

### Catagorical Independent Variable

```python
# origin is a categocal variable
# convert origin => into catagorical columns (create multiple columns from one columns)
cData = pd.get_dummies(cData, columns=['origin'])
cData.head()
```

```python
# it will calculate std, mean, count etc
cData.describe()
```

```python
cData.dtypes
```

```python
# now we have to consider on horsepower
hpIsDigit = pd.DataFrame(cData.horsepower.str.isdigit())
cData[hpIsDigit['horsepower'] == False]
```

```python
cData.median()
```

```python
cData = cData.replace('?', np.nan)
cData[hpIsDigit['horsepower'] == False]
```

```python
cData.median()
```

```python
medianFiller = lambda x: x.fillna(x.median())
cData = cData.apply(medianFiller,axis=0)

cData['horsepower'] = cData['horsepower'].astype('float64')
cData['horsepower'].head(10)
```
