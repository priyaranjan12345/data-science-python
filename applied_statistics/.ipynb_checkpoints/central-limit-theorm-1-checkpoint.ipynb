# Central Limit Theorm

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sample_size will be 10% of no.of_sample
def central_limit_theorm(data, n_sample=1000, sample_size=500, min_value=0, max_value=1338):
    print("cental limit theorm")
    """hello"""
    # impl central limit theorm
    b = {}
    
    for i in range (n_sample):
        x = np.unique(np.random.randint(min_value, max_value, size = sample_size))
        b[i] = data[x].mean()
        
    c = pd.DataFrame()
    c['Sample'] = b.keys()
    c['Mean'] = b.values()
    
    plt.figure(figsize = (15, 5))
    
    plt.subplot(1, 2, 1)
    sns.distplot(c.Mean)
    plt.title('hey')
    plt.xlabel('X')
    plt.ylabel('Y')
    
    plt.subplot(1, 2, 2)
    sns.distplot(data)
    plt.title('data')
    plt.xlabel('X')
    plt.ylabel('Y')
    
    plt.show()
```

```python
central_limit_theorm(charges, n_sample=5000, sample_size=500)
```
